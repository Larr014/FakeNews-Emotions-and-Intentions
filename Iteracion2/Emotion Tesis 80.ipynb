```
def emotionization(post, intensity):
  vector = []
  words = post.split(' ')
  
  for word in words:
    
    vector.append(word)
    for j in range(0,len(emotionVector)):

           
      if word[0] >= emotionVector['word'][j][0]:
        if word == emotionVector['word'][j] and emotionVector['emotion-intensity-score'][j]>=intensity :
          vector.append(emotionVector['emotion'][j])
      else:
        break
    if word=="rumor":
        return vector
  #df = pd.DataFrame(vector)
  return vector#df
```
