```
def get_tweet_by_id(id):
    tweet = api.get_status(id)
    return tweet.text
```

```
def search_tweet_by_id(fileName,df,columName,label):
  tweets = []
  for i in range(497,len(df[columName])):
    print(f'{(i*100)/len(df[columName])}')
    tweets.clear();
    try:
      dfSplit = df[columName][i].split()
      lTweets = len(dfSplit)
      for j in range(lTweets):
        print(f"\t {(j*100)/lTweets}")
        try:
          tweets.append(get_tweet_by_id(dfSplit[j]))
        except:
          pass
      
         
    #print(tweets)
      data = {'Tweet':  tweets,
           'label': label,
         
            }
      dfr = pd.DataFrame(data,columns=['Tweet',"label"])
      dfr.to_csv('/content/gdrive/My Drive/Colab Notebooks/dataset/tesis/fakeNewsNet/'+fileName+'_real/'+fileName+str(i)+'.csv') 
    except Exception as e:
      print("Exception: "+str(e))
      pass
    
    
  return tweets
```
