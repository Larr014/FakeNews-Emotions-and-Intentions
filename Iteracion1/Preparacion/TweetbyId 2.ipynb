```
def unirDataframes(nombreCarpeta,nombreArchivo,tam):
  df=pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/dataset/tesis/fakeNewsNet/'+nombreCarpeta+'/'+nombreArchivo+'0.csv')
  for i in range(1,tam):
    try:
      with open('/content/gdrive/My Drive/Colab Notebooks/dataset/tesis/fakeNewsNet/'+nombreArchivo+'noFake_convertido.csv', 'a') as f:
        df.to_csv(f)
      df=pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/dataset/tesis/fakeNewsNet/'+nombreCarpeta+'/'+nombreArchivo+str(i)+'.csv')
      #df = df.append(df2)
    except:
      pass
      
    if i%500==0:
        print(f'{(i*100)/tam}') 
  return df
```
