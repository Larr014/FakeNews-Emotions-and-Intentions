```
c = c.dropna()
```

```
fake = c[c[5]== "fake" ]
noFake = c[c[5] == "noFake"]
```

```
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```
def tokenization(text):
  tokens = word_tokenize(text)
  return tokens
```

```
fake.reset_index(inplace=True)
fake=fake.drop('index',axis=1)
fake=fake.drop(0,axis=1)
fake=fake.drop(1,axis=1)
fake
```

```
noFake.reset_index(inplace=True)
noFake=noFake.drop('index',axis=1)
noFake=noFake.drop(0,axis=1)
noFake=noFake.drop(1,axis=1)
noFake
```

```
for x in range(len(fake[2])):
  fake.loc[x][2] = tokenization(fake.loc[x][2])
  fake.loc[x][3] = tokenization(fake.loc[x][3])
  fake.loc[x][4] = tokenization(fake.loc[x][4])
```

```
transactions40 = fake[2].values.tolist()
transactions60 = fake[3].values.tolist()
transactions80 = fake[4].values.tolist()
```

```
!pip install apyori
import apyori
from apyori import apriori as apriori2
```

```
%time rules40 = apriori2(transactions40, min_support = 0.005, min_confidence = 0.8,  max_length=5)
%time rules60 = apriori2(transactions60, min_support = 0.005, min_confidence = 0.8,  max_length=5)
%time rules80 = apriori2(transactions80, min_support = 0.005, min_confidence = 0.8,  max_length=5)
```

```
%time results40 = list(rules40)
%time results60 = list(rules60)
%time results80 = list(rules80)
results40
```

```
def inspect(results):
    rh          = [tuple(result[2][0][0]) for result in results]
    lh          = [tuple(result[2][0][1]) for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(rh, lh, supports, confidences, lifts))
```

```
resultDataFrame40=pd.DataFrame(inspect(results40),
                columns=['rhs','lhs','support','confidence','lift'])
resultDataFrame60=pd.DataFrame(inspect(results60),
                columns=['rhs','lhs','support','confidence','lift'])
resultDataFrame80=pd.DataFrame(inspect(results80),
                columns=['rhs','lhs','support','confidence','lift'])
```
